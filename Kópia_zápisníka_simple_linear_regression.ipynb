# Simple Linear Regression

## Importing the libraries

```
import numpy
import pandas
import matplotlib.pyplot as plb
from sklearn.model_selection import train_test_split
```

## Importing the dataset

```
dataset = pandas.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

## Splitting the dataset into the Training set and Test set

```
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

## Training the Simple Linear Regression model on the Training set

```
from sklearn.linear_model import LinearRegression
lnr = LinearRegression()
lnr.fit(X_train, y_train)
```

## Predicting the Test set results

```
y_pred = lnr.predict(X_test)
```

## Visualising the Training set results

```
plb.scatter(X_train, y_train, color='red')
plb.plot(X_train, lnr.predict(X_train), color='blue')
plb.title('Salary vs Experience (Training set)')
plb.xlabel('Years of experience')
plb.ylabel('Salary')
plb.show()
```

## Visualising the Test set results

```
plb.scatter(X_test, y_test, color='red')
plb.plot(X_train, lnr.predict(X_train), color='blue')
plb.title('Salary vs Experience (Test set)')
plb.xlabel('Years of experience')
plb.ylabel('Salary')
plb.show()
```
